## Import Data

```python
flights = sns.load_dataset("flights")
```

## Dataset information

```python
print(flights)
```

### This categorical graph is plotting passenger count by month, from 0 - 600 range.

  1. Although the years this data is from is 1949 - 1960, the beginning of commercial flights, I was surprised to see the trend of busy months to slower months is similar to trends from today. Summer months tend to be busier as a whole, with winter being a slight tick up.
  2. Being the years that are tracked are at the beginning of commercial flights, I believe that's why the passenger count is such a low range, with numbers being close to 100 for a given month/ year.

```python
sns.catplot(data=flights, kind="swarm", x="month", y="passengers")
```

### In this statistical graph, I have it broke down by year, displaying the passgers flown during the first 11 years of commercial flight. 

  1. As you can see, in 1949 during the first year of commercial flights the amount is 100 - 150 passengers flown. Looking over the next several years you can see a dramatic incline in passengers flying commericially. You can see although the numbers were low, it was an emerging industry and tickets weren't easily accessible given fewer flights.

```python
sns.lmplot(data=flights, x="year", y="passengers")
```
